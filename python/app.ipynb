```python
l = []
while(True):
    print("""
    1. add element
    2. remove element
    3.display first three elements
    4. Display last three elements
    5. print all elements
    6. Exit""")
    ch = int(input("enter your choice"))
    if ch == 1:
        e = (input("enter element to add: "))
        l.append(e)
    elif ch == 2:
        if l.count() == 0:
            print("list is empty")
        else:
            l.pop()
    elif ch == 3:
        print(l[:3])
    elif ch == 4:
        if l.count() == 0:
            print("list is empty")
        else:
            print(l[-3:-1])
    elif ch ==5:
        print(l)
    elif ch ==6:
        print("exiting the program")
        break
    else:
        print("invalid choice")           
```
