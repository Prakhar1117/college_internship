```python
# NumPy in python--Numerical Python
import numpy as np
print(np.__version__)

# 1_D array in numpy
a=np.array([1,2,3,4,4])
print(a,type(a),a.dtype)

# 2-D array in numpy
a=np.array([[1,2],[3,4],[5,6]])
print("\n",a)

b=np.array(['ab','b','c'])  #U(n)-unicode(length) 
print(b,type(b),b.dtype)

# zero array
print(np.zeros(4)) # 1-D
print(np.zeros((3,4))) # 2-D

# array of ones
print("\n",np.ones((2,3)))

# Fortron ordered array of ones- columns are major and rows are minor
print("\n",np.ones((3,5),order='F'))

print("\n",np.arange(5,50,5))  # gives an array of specific range(start,end-not included,interval)

print(np.linspace(1,10))



```
