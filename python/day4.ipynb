# Functions:

```python
def function():
    print("Hello, World!")
function()    
```

```python
def add(a,b):
        return a+b
print(add(10,20))
```

```python
def place(city):
    print(f"The city you entered is: {city}")
place("New York")    
```

```python
def place(*city):# arbitraty positional arguments
    print(city)
place("New York", "Los Angeles", "Chicago")    
```

```python
def function(city3,city2,city1):
    print(f"City 1: {city1}, City 2: {city2}, City 3: {city3}")
function(city1="New York", city2="Los Angeles", city3="Chicago")    
```

```python
def function(**city): #arbitrary keyword arguments
    print(city)
function(city1="New York", city2="Los Angeles", city3="Chicago")    
```

```python
def function(place = "Unknown"):
    print(f"The place you entered is: {place}")
function()    
function("New York")    
```

```python
def function(fruit):

    for x in fruit:
        print(x)

function(["apple", "banana", "cherry"])        
```

# recursion:

```python
def rec(k):
    if k>0:
        print(k)
        k = k-1
        rec(k)
    else:
        print(k)
rec(5)            
```
