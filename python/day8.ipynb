```python
import pandas as pd

df = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
print(df)
print("\n addition",df+5)
print("\n subtraction",df-5)
print("\n multiplication",df*5)
print("\n division",df/5)
print("\n modulus",df%5)
print("\n floor division",df//5)


df.loc[2] = df.loc[2] + 5

print("--- Adding 5 to Row 2 Only ---")
print(df)
```

```python
import pandas as pd

df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[10,20,30,40,50]})
df2 = pd.DataFrame({"A":[10,20,30],"B":[400,500,600]}, index = [1,2,3])
print(df1)
print(df2)

print("\n addition\n",df1+df2)
print("\n subtraction\n",df1-df2)
print("\n multiplication\n",df1*df2)
print("\n division\n",df1/df2)
print("\n floor division\n",df1//df2)
```

```python
import pandas as pd

one = pd.DataFrame({
    "Name":["A1","B1","C1","D1"],
    "subjects":["Sub1","Sub2","Sub4","Sub6"],
    "Marks":[78,90,80,50]
},index = [1,2,3,4])


two = pd.DataFrame({
    "Name":["B1","C1","P1","Q1"],
    "subjects":["Sub1","Sub2","Sub3","Sub4"],
    "Marks":[78,90,80,50]
},index =[1,2,3,4])


result = pd.concat([one,two])
print(result)

result2 = pd.concat([one,two],keys=["x:","y:"])
print(result2)
```

```python
import pandas as pd

one = pd.DataFrame({
    "Id":[1,2,3,4],
    "Name":["A1","A2","A3","A4"],
    "subjects":["Sub1","Sub2","Sub4","Sub6"],
})


two = pd.DataFrame({
    "Id":[2,1,3,4],
    "Name":["B1","B2","B3","B4"],
    "subjects":["Sub1","Sub2","Sub3","Sub4"],
})


result = one.merge(two,on="Id")
print(result)
```
