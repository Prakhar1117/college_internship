```python
# CSV files
import csv
d=[
    ["Name","email","mobile"],
    ["prakhar","abcd@email.com","123456789"],
    ["priyanshu","abc@email.com","123456789"],
    ["shailja","abce@email.com","123456789"],
    ["hemant","abcf@email.com","123456789"] 
]

with open("data.csv","w",newline="")as file:
    writer=csv.writer(file)

    for row in d:
        writer.writerow(row)
```

```python
# exception handling
# zero error
try:
 n=int(input("enter a number"))
 print(n)
 print(10/n)
except ZeroDivisionError:
   print("divide by zero is not possible")
except ValueError:
  print("invalid number")
```

```python
# index error
try:
    i=int(input("enter index value"))
    a=[1,2,3,4,5,6]
    print(a[i])
except IndexError:
    print("invalid index")
```
