Q2 Explore more datetime function and uses in pandas

```python
import pandas as pd

# Sample DataFrame with date strings
df = pd.DataFrame({
    'date_strings': ['2026-05-01', '31/05/2026', '2026.06.15'],
    'sales': [100, 150, 200]
})

# Convert to datetime (format='mixed' handles multiple string formats automatically)
df['cleaned_date'] = pd.to_datetime(df['date_strings'], format='mixed')
print(df.dtypes)


# Assuming 'cleaned_date' is already converted
df['year'] = df['cleaned_date'].dt.year
df['month'] = df['cleaned_date'].dt.month
df['day_name'] = df['cleaned_date'].dt.day_name()
df['is_leap_year'] = df['cleaned_date'].dt.is_leap_year

print(df[['cleaned_date', 'year', 'day_name', 'is_leap_year']])
```
