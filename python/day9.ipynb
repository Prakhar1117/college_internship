```python
import pandas as pd

df = pd.DataFrame({
    "col1":range(12),
    "col2":["A"]*3+["B"]*3+["C"]*3 + ["D"]*3,
    "Date":pd.to_datetime(["2026-05-21","2026-05-22","2026-05-23"]*4)
})
print(df)
pivoted = df.pivot(index ="Date",columns="col2",values="col1")
print(pivoted)
```

```python
import pandas as pd

data={
    "Course":["CS","Civil","CS","Civil","CS"],
    "Year":[1,2,1,1,2],
    "Student":[100,150,120,90,180]
}
df = pd.DataFrame(data)
pivoted= df.pivot_table(index ="Course",columns="Year",values="Student",aggfunc="sum")
print(pivoted)
```

```python
import pandas as pd
import numpy as np

data = {"col1":[2,np.nan,np.nan,4],
        "col2":[1,pd.NA,pd.NA,2],
        
        }

df =pd.DataFrame(data)
print(df)

f= df.fillna("-")
print(f)
```

```python
import pandas as pd
import numpy as np

data = {"col1":[2,np.nan,np.nan,4],
        "col2":[1,pd.NA,pd.NA,2],
        
        }

df =pd.DataFrame(data)
print(df)
res = df.ffill()
print(res)

res2 = df.bfill()
print(res2)
```

```python
import pandas as pd
import numpy as np

data ={"one":[1,2,3,],
        "two":[5,6,7],
        "three":[8,9,0]}

df =pd.DataFrame(data, index=["a","c","d"],columns=["one","two","three"])
df = df.reindex(["a","b","c","d","e"])
print(df)

df1 = df.ffill()
print("\n",df1)

df2 = df.bfill()
print("\n",df2)
```
